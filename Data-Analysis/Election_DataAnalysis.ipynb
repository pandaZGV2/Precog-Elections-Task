# PreCog Election Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./All_States_GE_CSV.csv')
df.fillna("", inplace=True)
```

```python
df.head()
```

```python
print("DataFrame Info:")
print(df.info())

print("\nDataFrame Summary Statistics:")
print(df.describe())
```

```python
filtered_df = df[((df['Year'] - 2009) % 5 == 0) & (df['Year'] >= 2009)]

display(filtered_df)

df = filtered_df.copy()

# df['Year'] = df['Year'].astype(str)
```

```python
df['Votes'] = df['Votes'].astype(int)
new_df = df[['Year', 'Party', 'Votes']].copy()
party_votes_all = new_df.groupby(['Year', 'Party'])['Votes'].sum().reset_index()

top_5_parties = party_votes_all.groupby('Year').apply(lambda group: group.nlargest(5, 'Votes')).reset_index(drop=True)

pivot_df = top_5_parties.pivot(index='Year', columns='Party', values='Votes')

pivot_df = pivot_df.fillna(0)

display(pivot_df)

pivot_df.plot(kind='bar')

plt.xlabel('Year')
plt.ylabel('Votes')
plt.title('Total Votes by Party and Year')

plt.show()
```

```python
total_votes_per_year = party_votes_all.groupby('Year')['Votes'].sum()

display(total_votes_per_year)

party_votes_all['Percentage'] = party_votes_all.apply(lambda row: (row['Votes'] / total_votes_per_year[row['Year']]) * 100, axis=1)

top_5_parties_percentage = party_votes_all.groupby('Year').apply(lambda group: group.nlargest(5, 'Percentage')).reset_index(drop=True)

display(top_5_parties_percentage.head())

pivot_df = top_5_parties_percentage.pivot(index='Year', columns='Party', values='Percentage')

pivot_df = pivot_df.fillna(0)

pivot_df.plot(kind='bar', stacked=True)

plt.xlabel('Year')
plt.ylabel('Votes')
plt.title('% Votes by Party and Year')

plt.show()

pivot_df.plot(kind='line')

plt.xlabel('Year')
plt.ylabel('Votes')
plt.title('% Votes by Party and Year')

plt.show()
```

```python
state = 'Karnataka'

state_df = df.copy()
state_df = state_df[state_df['State_Name'] == state]
state_df = state_df[['Year', 'Party', 'Votes']]

party_votes_state = state_df.groupby(['Year', 'Party'])['Votes'].sum().reset_index()

total_votes_per_year = party_votes_state.groupby('Year')['Votes'].sum()

display(total_votes_per_year)

party_votes_state['Percentage'] = party_votes_state.apply(lambda row: (row['Votes'] / total_votes_per_year[row['Year']]) * 100, axis=1)

top_5_parties_percentage = party_votes_state.groupby('Year').apply(lambda group: group.nlargest(5, 'Percentage')).reset_index(drop=True)

display(top_5_parties_percentage)

pivot_df = top_5_parties_percentage.pivot(index='Year', columns='Party', values='Percentage')

pivot_df = pivot_df.fillna(0)

pivot_df.plot(kind='line')

plt.xlabel('Year')
plt.ylabel('Votes')
plt.title(f'% Votes by Party and Year for {state}')

plt.show()
```

```python
def winner_by_state_and_const(row):
    state = row['State_Name']
    constituency = row['Constituency_No']
    vote_share = row['Vote_Share_Percentage']
    year = row['Year']
    df[(df['State_Name'] == state) & (df['Constituency_No'] == constituency)]
    # display(df[(df['State_Name'] == state) & (df['Constituency_No'] == constituency)])
    max_vote_share = df[(df['State_Name'] == state) & (df['Constituency_No'] == constituency) & (df['Year'] == year)]['Vote_Share_Percentage'].max()
    # print(vote_share, max_vote_share)
    return bool(vote_share == max_vote_share)
```

```python
cand_df = df[df['Year'] == 2014].copy()
display(cand_df['Year'].tail())
cand_df['Won'] = cand_df.apply(winner_by_state_and_const, axis=1)
cand_df = cand_df[['State_Name', 'Sex', 'Party', 'Constituency_Type', 'Party_Type_TCPD', 'Turncoat', 'Incumbent', 'Recontest', 'MyNeta_education', 'Won']]
cand_df.dropna(inplace=True)
cand_df = cand_df[cand_df['Won'] == 1]

education_data = cand_df['MyNeta_education']
education_counter = education_data.value_counts()

education_counter = education_counter.drop('')
education_counter.plot(kind='bar')

plt.xlabel('Education')
plt.ylabel('Frequency')
plt.title('Frequency of Education in 2014 for winners')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Insights

* Between 2009 and 2019, the BJP experienced a consistent increase in voteshare across most states and the country.
* North-eastern states such as Nagaland, Tripura, and Manipur witnessed the highest turnout from 2009 to 2019, while states like Jammu and Kashmir, Bihar, and Uttar Pradesh had the lowest turnout during the same period.
* When analyzing the education of candidates, it was observed that a significant number held graduate and post-graduate degrees. However, there was also a considerable presence of less educated candidates. Interestingly, an examination of winning candidates showed a trend where the frequency of victory was in line with the level of education, albeit with few exceptions. This suggests that education may play a substantial role in determining voteshare.
  * Unfortunately, a deeper analysis involving factors like population and economic data per constituency was hindered due to insufficient data.
* Correlations with voter turnout by state:
  * Literacy rate showed a weak positive correlation, indicating that higher literacy rates may lead to increased voter participation. 
  * Sex ratio did not exhibit a statistically significant correlation with voter turnout.
  * GDP, however, displayed a weak but statistically significant correlation with voter turnout, suggesting that a decrease in a state's GDP might motivate more people to cast their votes.
